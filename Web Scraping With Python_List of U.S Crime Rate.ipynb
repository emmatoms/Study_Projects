```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_violent_crime_rate#Rate_by_year" 
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html")
```

```python
print(soup)
```

```python
table = soup.find_all("table")[0]
table
```

```python
crime_title = table.find_all("th")
crime_title
```

```python
crime_table_title =[title.text.strip() for title in crime_title]
crime_table_title
```

```python
import pandas as pd

df = pd.DataFrame (columns =crime_table_title)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```
